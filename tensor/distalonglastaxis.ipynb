```python
import tensorflow as tf
```

```python
shape=[32,10,100,3] #Tetszőleges shapekre is működik, nyugodtan lehet variálni

x1 = tf.random.uniform(shape,minval=-5,maxval=5,dtype=tf.double)
x2 = tf.random.uniform(shape,minval=-5,maxval=5,dtype=tf.double)
```

```python
#A transzponáláshoz a tengelyek felcseréléseinek megcsinálása (lehet lehetne elegánsabban de hajnal 2 van)
perm1=list(range(len(shape)+1))
perm1[-1]=len(shape)-1
perm1[-2]=len(shape)
perm2=list(range(len(shape)+1))
perm2[-1]=len(shape)-1
perm2[-2]=len(shape)-2
perm2[-3]=len(shape)
#Gyakorlatilag leduplikálja saz utolsó vektoroka egy új tengelyre, ami ugyanúgy az utolsó előtti axis hosszú
#gyakorlatilag generál egy [i,j,...,k,l,m] shape-ű A tenzorból egy [i,j,...,k,l,l,m] shapeű A'-t, ahol
#A[i,j,...,k,l,m]=#A'[i,j,...,k,l,p,m], minden p-re (itt most ezek ténylegesen indexek).
x1m=tf.transpose(tf.reshape(tf.repeat(x1,shape[-2]),shape+[shape[-2]]),perm1)
#Hasonlóan a  másikhoz, csak most B[i,j,...,k,l,m]=#B'[i,j,...,k,p,l,m], minden p-re
x2m=tf.transpose(tf.reshape(tf.repeat(x2,shape[-2]),shape+[shape[-2]]),perm2)
#Itt most (A'-B')[i,j,...,k,l,m,:]=A[i,j,...,k,l,:]-B[i,j,...,k,m,:] Ahol : azt jelenti hogy minden indexre igaz ott.
```
